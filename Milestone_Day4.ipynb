```python
CountOfCompany = ES['Company Name'].nunique()
```

```python
CountOfCompany #How many different companies are represented in the data set?
```

```python
Total_Job_Added = ES[ES['Borough'] =='Queens']['Job created'].sum()
```

```python
Total_Job_Added #What is the total number of jobs created for businesses in Queens?
```

```python
ES['Email_Domains'] = ES['company email'].str.split('@', n =1, expand = True)[1]
```

```python
Total_unique_EmailDomains = ES['Email_Domains'].nunique()
```

```python
Total_unique_EmailDomains #How many different unique email domains names are there in the data set?
```

```python
NTA_count = ES.groupby('NTA')['Company Name'].nunique()
```

```python
NTA_count
```

```python
Qualified_NTA = ES[ES['NTA'].isin(NTA_count[NTA_count>=5].index.tolist())]
```

```python
Qualified_NTA
```

```python
Qualified_NTA.groupby('NTA').mean()['Total Savings'] #average total savings for each NTA
```

```python
Qualified_NTA.groupby('NTA').sum()['Job created'] # total jobs created for each NTA
```
